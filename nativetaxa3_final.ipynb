```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = r'C:\myfilepath\Statewide_Terrestrial_Biodiversity_Summary_-_ACE_[ds1331].csv'
data = pd.read_csv(file_path)
```

```python
# Relevant columns for analysis
columns_of_interest = [
    'County', 'NtvAmph', 'RarAmph', 'NtvRept', 'RarRept', 'NtvBird', 'RarBird', 'NtvMamm', 'RarMamm'
]
```

```python
# Filter and sum relevant columns
data = data[columns_of_interest]
data['Total_Amphibians'] = data['NtvAmph'] + data['RarAmph']
data['Total_Reptiles'] = data['NtvRept'] + data['RarRept']
data['Total_Birds'] = data['NtvBird'] + data['RarBird']
data['Total_Mammals'] = data['NtvMamm'] + data['RarMamm']
```

```python
# Group by County and aggregate
grouped_data = data.groupby('County').sum().reset_index()
```

```python
# Save processed data
processed_data_path = 'C:/myfilepath/processed_data.csv'
grouped_data.to_csv(processed_data_path, index=False)
```

```python
# Load the processed data
processed_data_path = 'C:/myfilepath/processed_data.csv'
data = pd.read_csv(processed_data_path)
```

```python
# Analysis and visualization function
def plot_biodiversity(category):
    plt.figure(figsize=(12, 8))
    data_sorted = data.sort_values(by=f'Total_{category}', ascending=False).head(10)
    plt.barh(data_sorted['County'], data_sorted[f'Total_{category}'], color='gray')
    plt.xlabel(f'Summed Native and Rare Taxa Counts')
    plt.ylabel('County')
    plt.title(f'Top 10 CA Counties by Summed Native and Rare {category}')
    plt.gca().invert_yaxis()
    plt.savefig(f'C:/myfilepath/plot.png')
    plt.show()
```

```python
# Plot top county for each category
categories = ['Amphibians', 'Reptiles', 'Birds', 'Mammals']
for category in categories:
    plot_biodiversity(category)
```

```python
# Load the processed data
processed_data_path = 'C:/myfilepath/processed_data2.csv'
data = pd.read_csv(processed_data_path)

# Analysis and visualization function
def plot_biodiversity(data, category, native_column):
    plt.figure(figsize=(12, 8))
    data_sorted = data.sort_values(by=native_column, ascending=False).head(10)
    plt.barh(data_sorted['County'], data_sorted[native_column], color='gray')
    plt.xlabel(f'Native {category}')
    plt.ylabel('County')
    plt.title(f'Top 10 Counties by Native {category}')
    plt.gca().invert_yaxis()
    plt.savefig(f'C:/myfilepath/{category.lower()}_native_plot.png')
    plt.show()

# Plot top 10 counties for each category
categories = {
    'Amphibians': 'NtvAmph',
    'Reptiles': 'NtvRept',
    'Birds': 'NtvBird',
    'Mammals': 'NtvMamm'
}

for category, native_column in categories.items():
    plot_biodiversity(data, category, native_column)
```

```python
# Load the processed data
processed_data_path = 'C:/myfilepath/processed_data2.1.csv'
data = pd.read_csv(processed_data_path)

# Function to generate and display tables for top 10 counties by native species counts
def display_top_10_table(data, category, native_column):
    # Sort and select top 10
    top_10_data = data.sort_values(by=native_column, ascending=False).head(10)
    
    # Display the table
    print(f'Top 10 Counties by Native {category}')
    print(top_10_data.to_string(index=False))
    print('\n' + '-'*80 + '\n')

# Categories and corresponding native columns
categories = {
    'Amphibians': 'NtvAmph',
    'Reptiles': 'NtvRept',
    'Birds': 'NtvBird',
    'Mammals': 'NtvMamm'
}

# Display tables for each category
for category, native_column in categories.items():
    display_top_10_table(data, category, native_column)
```
